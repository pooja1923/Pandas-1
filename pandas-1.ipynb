# Ranking data

```python
e={
    'country':['India','China','Japan','Bangladesh','Sinpgapore'],
    'gold':[12,20,11,3,None]
}
f = pd.DataFrame(e)
f
```

# Mask filtering

```python
a = pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/covid_19_data.csv')
a.head()
```

```python
a.info()
```

```python
#convert ObservationDate,Last Update to date time

a['Last Update'] = pd.to_datetime(a['Last Update'])
a['ObservationDate'] = pd.to_datetime(a['ObservationDate'])
```

```python
a.info()
```

```python
#only printing results with country India

mask1 = a['Country/Region'] == 'India'
a[mask1]
```

```python
mask2 = a['ObservationDate'] == '2021-05-29'
a[mask1 & mask2]  
```

# isin()

```python
mask = a['Country/Region'].isin(['India',"Japan"])
a[mask]
```

**without isin()**

```python
mask3 = a['Country/Region'] == 'India'
mask4 = a['Country/Region'] == 'Japan'
a[mask3 | mask4]
```
