# Remame index

**1. Mapper + Axis**
**2. Index + Column**

**1. Mapper + Axis**

```python
d = pd.read_csv('/kaggle/input/data-science-jobs-in-india/naukri_data_science_jobs_india.csv',index_col=['Job_Role'])
d
```

```python
d.rename(mapper = {"Senior Data Scientist":"Senior Data Analyst",
                   "Data Scientist":"Data Analyst"},axis = 0,inplace = True)  #axis = 0/"index"/"rows"  means same
d
```

**2. Index + Column**

```python
d.rename(index = {"Senior Data Analyst":"Senior Data Scientist",
                  "Data Analyst":"Data Scientist"},inplace = True)
d
```

**Column Renaming**

```python
d.rename(mapper = {"Company":"Firm","Skills/Description":"Requirements"},axis = 1,inplace = True)
d
```

```python
d.rename(columns = {"Firm":"Company","Requirements":"Skills"},inplace = True)
d
```

**on dataframe**

```python
import numpy as np

a = pd.read_csv("/kaggle/input/data-science-jobs-in-india/naukri_data_science_jobs_india.csv")
a.head()
```

```python
a["Salary"] = np.random.randint(400000,4000000,12000)
a
```

```python
a["Salary"] = a["Salary"].apply(lambda x: "Rs. " + str(x))
```

```python
a
```

```python
def band(x):
    if x>2000000:
        return "A"
    elif x>1000000:
        return "B"
    else:
        return "C"
```

```python
a["Category"] = a["Salary"].apply(band)
```

```python
a
```

```python
a["Category"].value_counts()
```

**apply()--> across rows**

```python
def flag(x):
    loc = x["Location"]
    exp = x["Job Experience"]
    if loc in ["Bangalore/Bengaluru","Bangalore","Bengaluru"] and exp == "3-5":
        return "Yes"
    else:
        return "No"
```

```python
a["Apply or not"] = a.apply(flag, axis = 1)
```

```python
a
```

```python
a["Apply or not"].value_counts()
```
